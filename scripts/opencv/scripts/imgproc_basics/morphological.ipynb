```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def show_img(img):
    fig = plt.figure(figsize=(15, 15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')

def load_img():
    img = np.zeros((600,600))
    img = cv2.putText(img,'ABCDE',org=(50,300),fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale=5,color=(255,255,255),thickness=20)
    return img

def load_bricks(view = True):
    img = load_img('bricks.jpg')
    if view:
        show_img(img)
    return img
```

```python
img = load_img()
show_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
kernel
```

```python
eroded_image = cv2.erode(img,kernel,iterations=4)
show_img(eroded_image)
```

```python
img = load_img()
white_noise =  np.random.randint(low=0, high=2, size=(600,600))
```

```python
white_noise = white_noise * 255
noise_image =  white_noise + img
show_img(noise_image)
```

```python
opening  =  cv2.morphologyEx(noise_image,cv2.MORPH_OPEN,kernel)
show_img(opening)
```

```python
img =  load_img()
black_noise = np.random.randint(0,2,size=(600,600))
black_noise  =  black_noise*-255

```

```python
black_noise_image =  black_noise + img
black_noise_image[black_noise_image== -255]=0
show_img(black_noise_image)
```

```python
closing  =  cv2.morphologyEx(black_noise_image,cv2.MORPH_CLOSE,kernel)
show_img(closing)
```

```python
img =  load_img()
gradient = cv2.morphologyEx(img,cv2.MORPH_GRADIENT,kernel)
show_img(gradient)
```
