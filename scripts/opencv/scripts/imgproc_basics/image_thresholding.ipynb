```python
def show_img(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
img = cv2.imread('../../data/crossword.jpg',0)
show_img(img)
```

```python
ret,thresh = cv2.threshold(img, 200,255,cv2.THRESH_BINARY)
show_img(thresh)
```

```python
thresh_ad = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,8)
show_img(thresh_ad)
```

```python
thesh_blended = cv2.addWeighted(thresh,0.8,thresh_ad,0.2,1)
show_img(thesh_blended)
```
