```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_image = np.zeros((512,512,3))
blank_canvas = blank_image.copy()
plt.imshow(blank_image)
```

```python
cv2.circle(img=blank_canvas,center=(255,255),radius=25,color=(0,255,0),thickness=-1)
plt.imshow(blank_canvas)
```

```python
cv2.rectangle(img=blank_canvas,pt1=(20,20),pt2=(100,100),color=(255,255,0),thickness=5)
plt.imshow(blank_canvas)
```

```python
cv2.line(img=blank_canvas,pt1=(0,0),pt2=(512,512),color=(102,102,5),thickness=5)
plt.imshow(blank_canvas)
```

```python
blank_canvas2 = blank_image.copy()
points = np.array([[100,200],[300,400],[420,100],[300,50]])
pts = points.reshape((-1,1,2))
pts.shape
pts
```

```python
cv2.polylines(img=blank_canvas2,pts=[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_canvas2)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(img=blank_canvas2,text='Hello World',org=(100,100),fontFace=font,fontScale=2,color=(102,102,102),thickness=4,lineType=cv2.LINE_AA)
plt.imshow(blank_canvas2)
```
